# **Linear Regression from Scratch**
In this tutorial notebook, we'll implement **Linear Regression** along with its optimization algorithms: **Gradient Descent** (Batch) and **Stochastic Gradient Descent (SGD)**.

---

## **1. Introduction to Linear Regression**
Linear Regression is a fundamental algorithm used for predicting a continuous target variable ($y$) based on one or more input features ($\mathbf{x}$).

### **1.1 Hypothesis Function (One-Dimensional)**
In the simplest case, when we have only one input feature, the hypothesis function models the relationship between input and output as:
$$
h_\theta(x) = \theta_0 + \theta_1 x
$$
Where:
- $x$: Input feature
- $h_\theta(x)$: Predicted value (output)
- $\theta_0$: Bias term
- $\theta_1$: Model parameter (weight)

If we include a bias term $x_0 = 1$, we can represent this in vectorized form as:
$$
h_\theta(\mathbf{x}) = \mathbf{x}^\top \boldsymbol{\theta}
$$
Where:
- $\mathbf{x} = [1, x]$: Input feature vector
- $\boldsymbol{\theta} = [\theta_0, \theta_1]^\top$: Parameter vector

The goal of linear regression is to find the parameters $\boldsymbol{\theta}$ that minimize the error between the predicted values $h_\theta(x)$ and the actual target values $y$.

---

## **2. Cost Function (One-Dimensional)**

To measure how well the model predicts the target, we use the **Mean Squared Error (MSE)** as the cost function:
$$
J(\boldsymbol{\theta}) = \frac{1}{2m} \sum_{i=1}^m \left( h_\theta(x^{(i)}) - y^{(i)} \right)^2
$$
Where:
- $m$: Number of training examples
- $h_\theta(x^{(i)})$: Predicted value for the $i$-th example
- $y^{(i)}$: Actual value for the $i$-th example

In vectorized form, this becomes:
$$
J(\boldsymbol{\theta}) = \frac{1}{2m} (\mathbf{X}\boldsymbol{\theta} - \mathbf{y})^\top(\mathbf{X}\boldsymbol{\theta} - \mathbf{y})
$$
Where:
- $\mathbf{X}$: Matrix of input features (size $(m \times 2)$ for one feature including bias)
- $\mathbf{y}$: Vector of target values
- $\boldsymbol{\theta}$: Parameter vector

---

## **3. Optimization with Gradient Descent (One-Dimensional)**

### **3.1 What is Gradient Descent?**
Gradient Descent is an iterative optimization algorithm used to minimize the cost function by adjusting the model parameters $\boldsymbol{\theta}$ step by step.

### **3.2 Algorithm Steps**
1. Initialize parameters $\boldsymbol{\theta}$ (typically to zeros or random values).
2. Repeat until convergence (defined as when the cost function change is below a threshold or after a fixed number of iterations):
   $$
   \theta_j := \theta_j - \alpha \frac{\partial J(\boldsymbol{\theta})}{\partial \theta_j}
   $$
   Where:
   - $\alpha$: Learning rate, controlling the size of the update step
   - $\frac{\partial J(\boldsymbol{\theta})}{\partial \theta_j}$: Partial derivative with respect to $\theta_j$

### **3.3 Gradient Calculation (One-Dimensional)**
The partial derivative of the cost function is:
$$
\frac{\partial J(\boldsymbol{\theta})}{\partial \theta_j} = \frac{1}{m} \sum_{i=1}^m \left( h_\theta(x^{(i)}) - y^{(i)} \right) x_j^{(i)}
$$
For one-dimensional data:
- $x_0^{(i)} = 1$ (bias term)
- $x_1^{(i)} = x^{(i)}$ (first input feature value)

---

## **4. Transition to Multi-Dimensional Data**

Now, let's extend our understanding to **multi-dimensional data**, where we have more than one input feature.

For multi-dimensional data:
- Input matrix $\mathbf{X}$ has shape $(m \times (n+1))$, including the bias term
- Parameter vector $\boldsymbol{\theta}$ has shape $((n+1) \times 1)$
- Output vector $\mathbf{y}$ has shape $(m \times 1)$

### **4.1 Hypothesis Function (Multi-Dimensional)**
When we have multiple features ($x_1, x_2, \dots, x_n$), the hypothesis function becomes:
$$
h_\theta(\mathbf{x}) = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \dots + \theta_n x_n
$$
In vectorized form, it is:
$$
h_\theta(\mathbf{x}) = \mathbf{x}^\top \boldsymbol{\theta}
$$
Where:
- $\mathbf{x} = [1, x_1, x_2, \dots, x_n]$: Input feature vector
- $\boldsymbol{\theta} = [\theta_0, \theta_1, \theta_2, \dots, \theta_n]^\top$: Parameter vector

---

## **5. Cost Function (Multi-Dimensional)**
For multi-dimensional data, the cost function remains the same, but we calculate it with matrix operations. The vectorized form is:
$$
J(\boldsymbol{\theta}) = \frac{1}{2m} (\mathbf{X}\boldsymbol{\theta} - \mathbf{y})^\top(\mathbf{X}\boldsymbol{\theta} - \mathbf{y})
$$

---

## **6. Optimization with Gradient Descent (Multi-Dimensional)**

### **6.1 Gradient Calculation (Multi-Dimensional)**
The gradient of the cost function with respect to each parameter is:
$$
\frac{\partial J(\boldsymbol{\theta})}{\partial \theta_j} = \frac{1}{m} \sum_{i=1}^m \left( h_\theta(\mathbf{x}^{(i)}) - y^{(i)} \right) x_j^{(i)}
$$

In vectorized form, the update rule becomes:
$$
\boldsymbol{\theta} := \boldsymbol{\theta} - \alpha \nabla J(\boldsymbol{\theta})
$$
Where:
$$
\nabla J(\boldsymbol{\theta}) = \frac{1}{m} \mathbf{X}^\top (\mathbf{X}\boldsymbol{\theta} - \mathbf{y})
$$

---

## **7. Stochastic Gradient Descent (SGD)**

### **7.1 What is SGD?**
Unlike Batch Gradient Descent, which uses all training examples for each update, SGD updates parameters using a single randomly selected example at a time.

### **7.2 Algorithm Steps**
1. Shuffle the dataset.
2. For each training example $(\mathbf{x}^{(i)}, y^{(i)})$:
   $$
   \boldsymbol{\theta} := \boldsymbol{\theta} - \alpha \left( h_\theta(\mathbf{x}^{(i)}) - y^{(i)} \right)\mathbf{x}^{(i)}
   $$

### **7.3 Comparison with Batch Gradient Descent**
- **Batch Gradient Descent**: Computes gradient using all $m$ examples per update.
- **Stochastic Gradient Descent**: Computes gradient using one example per update.
- **Mini-batch Gradient Descent**: Computes gradient using a batch of $b$ examples per update ($b < m$).

```python
import numpy as np 
import matplotlib.pyplot as plt 
from typing import Tuple, Optional 
```

```python

class LinearRegression:
    def __init__(self, learning_rate: float = 0.01, max_iterations: int = 1000,
                 tolerance: float = 1e-6):
        """
        Initialize Linear Regression model.
        
        Args:
            learning_rate: Learning rate for gradient descent, usually set to 0.01 in theory
            max_iterations: Maximum number of iterations for optimization
            tolerance: Convergence criterion for optimization

        `tolerance` is the minimum difference between our old and new parameter value needed to consider Gradient Descent has converged and it is usually set to a small value like epsilon  
        """
        self.learning_rate = learning_rate
        self.max_iterations = max_iterations
        self.tolerance = tolerance
        self.theta = None
        self.cost_history = []
        
    def add_bias(self, X: np.ndarray) -> np.ndarray:
        """Add bias term to input features."""
        if len(X.shape) == 1:
            X = X.reshape(-1, 1)
        return np.c_[np.ones((X.shape[0], 1)), X]
    
    def hypothesis(self, X: np.ndarray) -> np.ndarray:
        """
        Calculate hypothesis function h_θ(x) = X θ
        
        Args:
            X: Input features (with bias term)
        Returns:
            Predicted values
        """
        return np.dot(X, self.theta.reshape(-1, 1))
    
    def compute_cost(self, X: np.ndarray, y: np.ndarray) -> float:
        """
        Compute Mean Squared Error cost function.
        
        Args:
            X: Input features (with bias term)
            y: Target values
        Returns:
            Cost value
        """
        m = X.shape[0]
        y = y.reshape(-1, 1)
        predictions = self.hypothesis(X)
        return (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    
    def compute_gradient(self, X: np.ndarray, y: np.ndarray) -> np.ndarray:
        """
        Compute gradient of the cost function.
        
        Args:
            X: Input features (with bias term)
            y: Target values
        Returns:
            Gradient vector
        """
        m = X.shape[0]
        y = y.reshape(-1, 1)
        predictions = self.hypothesis(X)
        return (1 / m) * np.dot(X.T, (predictions - y))
    
    def fit_batch_gd(self, X: np.ndarray, y: np.ndarray) -> None:
        """
        Fit model using batch gradient descent.
        
        Args:
            X: Input features
            y: Target values
        """
        # Add bias term and ensure proper shapes
        X = self.add_bias(X)
        y = y.reshape(-1, 1)
        m, n = X.shape
        
        # Initialize parameters
        self.theta = np.zeros((n, 1))
        
        # Gradient descent
        for i in range(self.max_iterations):
            old_theta = self.theta.copy()
            
            # Compute gradient and update parameters
            gradient = self.compute_gradient(X, y)
            self.theta -= self.learning_rate * gradient
            
            # Compute and store cost
            cost = self.compute_cost(X, y)
            self.cost_history.append(cost)
            
            # Check convergence
            if np.all(np.abs(self.theta - old_theta) < self.tolerance):
                break
    
    def fit_sgd(self, X: np.ndarray, y: np.ndarray, batch_size: int = 1) -> None:
        """
        Fit model using stochastic gradient descent.
        
        Args:
            X: Input features
            y: Target values
            batch_size: Size of mini-batch (1 for pure SGD)
        """
        # Add bias term and ensure proper shapes
        X = self.add_bias(X)
        y = y.reshape(-1, 1)
        m, n = X.shape
        
        # Initialize parameters
        self.theta = np.zeros((n, 1))
        
        # SGD
        for i in range(self.max_iterations):
            old_theta = self.theta.copy()
            cost = 0
            
            # Shuffle data
            indices = np.random.permutation(m)
            X_shuffled = X[indices]
            y_shuffled = y[indices]
            
            # Mini-batch updates
            for j in range(0, m, batch_size):
                X_batch = X_shuffled[j:j + batch_size]
                y_batch = y_shuffled[j:j + batch_size]
                
                # Compute gradient and update parameters
                gradient = self.compute_gradient(X_batch, y_batch)
                self.theta -= self.learning_rate * gradient
                
            # Compute and store cost for entire dataset
            cost = self.compute_cost(X, y)
            self.cost_history.append(cost)
            
            # Check convergence
            if np.all(np.abs(self.theta - old_theta) < self.tolerance):
                break
    # The normal equation is, a way of getting the best value for our parameters without the need for multiple iterations 
    def normal_equation(self, X: np.ndarray, y: np.ndarray) -> None:
        """
        Fit model using the normal equation.
        
        Args:
            X: Input features
            y: Target values
        """
        X = self.add_bias(X)
        y = y.reshape(-1, 1)
        self.theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
    
    def predict(self, X: np.ndarray) -> np.ndarray:
        """
        Make predictions for input features.
        
        Args:
            X: Input features
        Returns:
            Predicted values
        """
        X = self.add_bias(X)
        return self.hypothesis(X)
```

```python
def generate_sample_data(n_samples: int = 100, noise: float = 10) -> Tuple[np.ndarray, np.ndarray]:
    """
    Generate sample data for testing.
    
    Args:
        n_samples: Number of samples to generate
        noise: Standard deviation of Gaussian noise
    Returns:
        Tuple of features X and target values y
    """
    X = np.linspace(0, 100, n_samples).reshape(-1, 1)
    y = 2 * X + 5 + np.random.normal(0, noise, (n_samples, 1))
    return X, y
```

```python
def plot_regression(X: np.ndarray, y: np.ndarray, model: LinearRegression,
                   title: Optional[str] = None) -> None:
    """
    Plot regression line and data points.
    
    Args:
        X: Input features
        y: Target values
        model: Fitted LinearRegression model
        title: Plot title
    """
    plt.figure(figsize=(10, 6))
    plt.scatter(X, y, color='blue', label='Data points')
    y_pred = model.predict(X)
    plt.plot(X, y_pred, color='red', label='Regression line')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.title(title or 'Linear Regression')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def plot_cost_history(cost_history: list, title: Optional[str] = None) -> None:
    """
    Plot cost history during training.
    
    Args:
        cost_history: List of cost values during training
        title: Plot title
    """
    plt.figure(figsize=(10, 6))
    plt.plot(cost_history)
    plt.xlabel('Iteration')
    plt.ylabel('Cost')
    plt.title(title or 'Cost History')
    plt.grid(True)
    plt.show()
```

```python
if __name__ == "__main__":
    # Generate sample data
    X, y = generate_sample_data(100, noise=10)
    
    # Create and fit model using batch gradient descent
    model_bgd = LinearRegression(learning_rate=0.0001, max_iterations=1000)
    model_bgd.fit_batch_gd(X, y)
    
    # Plot results
    plot_regression(X, y, model_bgd, "Batch Gradient Descent")
    plot_cost_history(model_bgd.cost_history, "Cost History - Batch GD")
    
    # Create and fit model using SGD
    model_sgd = LinearRegression(learning_rate=0.0001, max_iterations=50)
    model_sgd.fit_sgd(X, y, batch_size=1) # Set batch_size to 1 for SDG, for mini-batc GD we can set it to any bacth size of our choice  
    
    # Plot results
    plot_regression(X, y, model_sgd, "Stochastic Gradient Descent")
    plot_cost_history(model_sgd.cost_history, "Cost History - SGD")
    
    # Fit using normal equation
    model_ne = LinearRegression()
    model_ne.normal_equation(X, y)
    
    # Plot results
    plot_regression(X, y, model_ne, "Normal Equation")
```
